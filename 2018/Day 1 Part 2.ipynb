```python
def duplicate_frequency():
    should_run = True
    count = 0
    seen = {}
    output = None
    runs = 0
    while should_run:
        runs += 1
        print(runs)
        file = open("/Users/tylerroth/code/adventofcode/2018/day1_data.txt")
        for line in file.readlines():
            function = line[0]
            print(count,seen, function, line[1:-1])
            if function == "+":
                count += int(line[1:-1])
            else: 
                count -= int(line[1:-1])
                
            if not count in seen:
                seen[count] = 1
            else:
                seen[count] += 1
                if seen[count] ==2: 
                    should_run = False
                    print("hit")

                    output = count
                    return seen
    return seen
```

```python
dupe = duplicate_frequency()
```
